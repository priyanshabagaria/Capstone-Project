```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
import dill
from collections import defaultdict
rest=dill.load(open('./data/NYC-new-restaurant.dill', 'rb'))
rest_google=dill.load(open('./data/restaurant-NYC-google-places.dill','rb'))
count=0
features=defaultdict(list)
for r in rest:
    if 'price' not in rest[r]:
        continue
    if r not in rest_google:
        continue
    features['business_id'].append(rest[r]['id'])
    categories=[]
    for eachCat in rest[r]['categories']:
        categories.append(eachCat['alias'])
    features['categories'].append(categories)
    count+=1
    #print(count)
    features['city'].append(rest[r]['location']['city'])
    features['user_ratings_total'].append(rest_google[r]['user_ratings_total'])
    features['coordinates.latitude'].append(rest[r]['coordinates']['latitude'])
    features['coordinates.longitude'].append(rest[r]['coordinates']['longitude'])
    features['name'].append(rest[r]['name'])
    features['display_phone'].append(rest[r]['display_phone'])
    features['location.display_address'].append(rest[r]['location']['display_address'])
    features['location.state'].append(rest[r]['location']['state'])
    features['location.zipcode'].append(rest[r]['location']['zip_code'])
    features['is_closed'].append(rest[r]['is_closed'])
    features['rating'].append(rest[r]['rating'])
    features['review_count'].append(rest[r]['review_count'])
    features['price'].append(rest[r]['price'])
    features['transcations'].append(rest[r]['transactions'])
    features['url'].append(rest[r]['url'])
df1=pd.DataFrame(features)
    
```

```python
dill.dump(df1,open('./data/NYC-features.dill','wb'))
```

```python
import dill
import json
from collections import defaultdict
import datetime as dt
import time
rest2_google=dill.load(open('./data/restaurant-NYC-google-review.dill', 'rb'))
print(rest2_google)
coun=0
revie=defaultdict(list)
for r in rest2_google:
    count+=1
    print(count)
    try:
        for review in rest2_google[r]['result']['reviews']:
            revie['business_id'].append(r)
            revie['text'].append(review['text'])
            revie['user_id'].append(review['author_name'])
            revie['relative_time_description'].append(review['relative_time_description'])
            d=dt.datetime.utcfromtimestamp(review['time']).strftime("%Y-%m-%d %H:%M:%S")
            revie['time_created'].append(d)
            revie['user_rating'].append(review['rating'])
    except:
        continue      
df2=pd.DataFrame(revie)
```
