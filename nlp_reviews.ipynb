```python
import string
import nltk
%matplotlib inline
from wordcloud.wordcloud import WordCloud, STOPWORDS
from PIL import Image
from nltk.corpus import stopwords
def text_clean(message):
    nopunc = [i for i in message if i not in string.punctuation]
    nn = "".join(nopunc)
    nn = nn.lower().split()
    nostop = [words for words in nn if words not in stopwords.words('english')]
    return(nostop)
```

```python
good = df_reviews[df_reviews['Customer EXP'] == 'GOOD']
bad = df_reviews[df_reviews['Customer EXP'] == 'BAD']
neu = df_reviews[df_reviews['Customer EXP'] == 'NEUTRAL']
```

```python
from nltk.corpus import stopwords
good_bow = text_clean(good['text'])
```

```python
bad_bow = text_clean(bad['text'])
```

```python
neu_bow = text_clean(neu['text'])
```

```python
good_para = ' '.join(good_bow)
bad_para = ' '.join(bad_bow)
new_para = ' '.join(neu_bow)
```

```python
stopwords = set(STOPWORDS)
stopwords.add('one')
stopwords.add('also')
stopwords.add('food')
mask_image = np.array(Image.open("cloud.jpg"))
wordcloud_good = WordCloud(colormap = "Paired",mask = mask_image, font_path = "./chint_.ttf", width = 300, height = 200, scale=2,max_words=1000, stopwords=stopwords).generate(good_para)
plt.figure(figsize = (7,10))
plt.imshow(wordcloud_good, interpolation="bilinear", cmap = plt.cm.autumn)
plt.axis('off')
plt.figure(figsize = (10,6))
plt.show()
wordcloud_good.to_file("good.png")
```
